```python
### Code - Leave-one-out cross validation
# https://www.statology.org/leave-one-out-cross-validation-in-python/
# https://medium.com/@svanillasun/how-to-deal-with-cross-validation-based-on-knn-algorithm-compute-auc-based-on-naive-bayes-ff4b8284cff4

from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score

#define cross-validation method to use
cv = LeaveOneOut()

#build knn model and fit with training data
model = KNeighborsClassifier(n_neighbors=3)
model.fit(feature_train_standardized, label_train)

#use LOOCV to evaluate model with testing data
scores = cross_val_score(model, feature_test_standardized, label_test, scoring='neg_mean_absolute_error',
                         cv=cv, n_jobs=-1)

#view mean absolute error
print(np.mean(np.absolute(scores)))
```

## <font color = forestgreen> 3. Model selection with leave-one-out cross validation

So far, we've trained one model at a time and I've given the value of k for you. Accuracy is what it is (no spoilers here), but could we still do a little better? Let's try that by <font color = darkorchid><b>hyperparameter tuning</b></font>.

As does the k-NN, I just happened to choose the k value of 3 for you. Repeat the model performance estimation with values $k=1...30$ using again leave-one-out cross validation.

```python
### Code - Select best k
x_range = range(1, 31)
mae = []
for x in x_range:
    import logging

    #build knn model
    model = KNeighborsClassifier(n_neighbors=x)
    model.fit(feature_train_standardized, label_train)

    #use LOOCV to evaluate model
    scores = cross_val_score(model, feature_test_standardized, label_test, scoring='neg_mean_absolute_error',
                         cv=cv, n_jobs=-1)
    #mean absolute error
    mae.append(np.mean(np.absolute(scores)))

#plot mean absolute error values
plt.plot(x_range, mae)
plt.xlabel('Value of k for KNN')
plt.ylabel('Cross-Validated mean absolute error')
plt.show()
```

```python
#knn
x_range = range(1, 31)
mae = []
for x in x_range:
    import logging

    #build knn model
    model = KNeighborsRegressor(n_neighbors=x)
    model.fit(boston_feature_train, boston_label_train)

    #use LOOCV to evaluate model
    scores = cross_val_score(model, boston_feature_test, boston_label_test, scoring='neg_mean_absolute_error',
                         cv=cv, n_jobs=-1)
    #mean absolute error
    mae.append(np.mean(np.absolute(scores)))

#plot mean absolute error values
print(mae)
plt.plot(x_range, mae)
plt.xlabel('Value of k for KNN')
plt.ylabel('Cross-Validated mean absolute error')
plt.show()
```

```python
# ridge regression
# https://machinelearningmastery.com/ridge-regression-with-python/
x_range = range(-10, 10)
ridge_mae = []
for x in x_range:
    # define model
    model = Ridge(alpha=2**x)
    model.fit(boston_feature_train, boston_label_train)

    # evaluate model
    scores = cross_val_score(model, boston_feature_test, boston_label_test, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
    
    #mean absolute error
    ridge_mae.append(np.mean(np.absolute(scores)))

#plot
print(ridge_mae)
plt.plot(x_range, ridge_mae)
plt.xlabel('lambda')
plt.ylabel('Cross-Validated mean absolute error')
plt.show()
```

```python
#predictions 
# from array it seems 2^(-5) was the best so lets use it for predictions
ridge_model = Ridge(alpha=2**(-5))

    
#fit training data to classifier
ridge_model.fit(boston_feature_train, boston_label_train)

#make predictions
ridge_prediction = ridge_model.predict(boston_feature_test)

#plot
plt.hist(ridge_prediction, bins=20)
plt.show()
```
